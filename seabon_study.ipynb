```python
import seaborn as sns

sns.get_dataset_names()
```

```python
penguins=sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.columns
```

```python
penguins["species"].value_counts()
```

```python
#scatter plot
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g")
# in the scatter plot we can conclude as flipper length increases body mass also increases
```

```python
#scatter plot with hue [group by ]
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="island")
# from this plot we can understand biscoe island penguine with high flipper length and bodymass
```

```python
#scatter plot with hue [group by sex ]
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
# from this plot we can understand male penguine with high flipper length and bodymass
```

```python
#sns.despine() remove top and right bar
#sn.despine(left=True) remove left
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
sns.despine()
# from this plot we can understand male penguine with high flipper length and bodymass
```

```python
#set_style and set_context
sns.set_style("darkgrid")#'white', 'dark', 'whitegrid', 'darkgrid', 'ticks' => grid color
sns.set_context("notebook") #'paper', 'notebook', 'talk', 'poster' => chart size
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python

sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="sex",palette="dark")
```

```python
#strip plot

sns.stripplot(data=penguins,x="species",y="body_mass_g",hue="island",dodge=True)
```

```python
#swarm plot
sns.set_context("talk")

sns.swarmplot(data=penguins,x="island",y="body_mass_g",hue="island")
```

```python
#histogram

sns.set_context("notebook")

sns.histplot(data=penguins,x="body_mass_g")
```

```python
sns.set_context("notebook")

sns.histplot(data=penguins,x="body_mass_g",hue="sex")
```

```python
#regression plot fits a regression line

sns.regplot(data=penguins,x="body_mass_g",y="flipper_length_mm",color="green")
```

```python
#line plot
sns.lineplot(data=penguins,x="body_mass_g",y="flipper_length_mm")
```

```python
sns.lineplot(data=penguins,x="body_mass_g",y="flipper_length_mm",hue="sex")
```

```python
#bar plot
sns.barplot(data=penguins,x="species",y="body_mass_g",hue="sex",palette=["pink","skyblue"])
```

```python
import numpy as np
sns.barplot(data=penguins,x="species",y="body_mass_g",hue="sex",palette=["pink","skyblue"],estimator=np.sum)
```

```python
#count plot

sns.countplot(data=penguins,x="species",hue="island",palette="Set2")
```
